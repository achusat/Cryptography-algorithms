```python
import random
```

```python
def Alice_public(a, public):
    A = (public["g"]**a) % public["n"]
    public["A"] = A
    return public
```

```python
def Bob_public(b, public):
    B = (public["g"]**b) % public["n"]
    public["B"] = B
    return public
```

```python
def Alice_secret(a, public):
    s = (public["B"]**a) % public["n"]
    return s
```

```python
def Bob_secret(b, public):
    s = (public["A"]**b) % public["n"]
    return s
```

```python
def initialize_public():
    public["g"] = random.randint(2,10)
    public["n"] = random.randint(2,1000)
    return public
```

```python
def initialize_private():
    private["a"] = random.randint(2,10)
    private["b"] = random.randint(2,20)
    return private
```

```python
public = dict()
public = initialize_public()
print("Randomly initialized modulus and base: " + str(public))
private = dict()
private = initialize_private()
print("Randomly initialized private keys: " + str(public))
public = Alice_public(private["a"], public)
public = Bob_public(private["b"], public)
print("Public variables after generation of public key: " + str(public))
print("Secret generated by Alice: " + str(Alice_secret(private["a"], public)))

print("Secret generated by Bob: " + str(Bob_secret(private["b"], public)))
```
